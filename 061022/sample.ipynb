```python
import pandas as pd
```

```python
uriage_data = pd.read_csv("C:/ubion/csv/uriage.csv")

uriage_data.head()
```

```python
# purchase_date 컬럼의 데이터 타입을 시계열로 변경

uriage_data["purchase_date"] = pd.to_datetime(uriage_data["purchase_date"])

uriage_data.dtypes
```

```python
uriage_data.head()
```

```python
# item_name 컬럼의 값들을 대문자로 변환

uriage_data["item_name"] = uriage_data["item_name"].str.upper()

uriage_data.head()
```

```python
# item_name 컬럼의 공백을 지운다.

uriage_data["item_name"] = uriage_data["item_name"].str.replace(" ", "")

uriage_data.head()
```

```python
# item_name 컬럼을 기준으로 오름차순 정렬

# uriage_data = uriage_data.sort_values("item_name", ascending=True)
uriage_data.sort_values("item_name", ascending=True, inplace=True)

uriage_data
```
