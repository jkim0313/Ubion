1. transaction으로 시작되는 파일 4개 불러오기
2. transaction_1, transaction_2 데이터프레임 결합(행을 추가)
3. transaction_detail_1, transaction_detail_2 데이터프레임 결합(행을 추가)
4. transaction, transaction_detail 데이터프레임 결합(공통인 컬럼의 값을 기준으로 열을 추가)

```python
import pandas as pd
```

```python
trans_1 = pd.read_csv("C:/ubion/csv/transaction_1.csv")
trans_2 = pd.read_csv("C:/ubion/csv/transaction_2.csv")
detail_1 = pd.read_csv("C:/ubion/csv/transaction_detail_1.csv")
detail_2 = pd.read_csv("C:/ubion/csv/transaction_detail_2.csv")
i_mast = pd.read_csv("C:/ubion/csv/item_master.csv")
c_mast = pd.read_csv("C:/ubion/csv/customer_master.csv")
```

```python
trans_1.head()
```

```python
trans_2.head()
```

```python
## 2. transaction_1과 transaction_2를 유니언 결합을 한다. 

trans_union = pd.concat([trans_1, trans_2], ignore_index=True)
trans_union
```

```python
## 3. transaction_detail_1과 transaction_detail_2를 유니언 결합을 한다. 

detail_dfs = [detail_1, detail_2]

detail_union = pd.concat(detail_dfs, ignore_index=True)
detail_union
```

```python
## 4. 2,3 번의 데이터프레임의 값을 조인 결합을 한다. 

td_join = pd.merge(detail_union, trans_union, how = "inner", on = "transaction_id")

td_join
```

```python
## 5. 4번 작업한 데이터프레임을 customer_master, item_master 와 조인 결합한다.

tdc_join = pd.merge(td_join, c_mast, how="inner", on="customer_id")

tdc_join
```
