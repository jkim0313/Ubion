```python
## left join transaction_detail_1, item_master 
## join 결합을 한 데이터에서 quantity * item_price 계산한 price 컬럼 추가
## 출력 컬럼은 transaction_id, quantity, item price, 
```

```python
sql = """ 
SELECT * from item_master
"""

_db.executeAll(sql)
```

```python
sql = """ 
SELECT 
    tr_d.`transaction_id`, tr_d.`quantity`, i_m.`item_price`,
    (tr_d.`quantity` * i_m.`item_price`) as price
FROM transaction_detail_1 as tr_d
    left join item_master as i_m
ON tr_d.`item_id` = i_m.`item_id` 
"""

_db.executeAll(sql)
```

1. transaction_1에 있는 payment_date = datetime 형태 
2. 파생변수(payment_month) = payment_date 값에서 년과 월만 따로 추출해서 삽입
3. 월별로 그룹화 price 평균, 합계 값을 출력

```python
sql = """ 
SELECT * FROM transaction_1
"""

_db.executeAll(sql)
```

```python
sql = """ 
SELECT DATE_FORMAT(`payment_date`, '%Y%m') as payment_month,
AVG(`price`) as AVG,
SUM(`price`) as SUM
FROM transaction_1 
GROUP BY payment_month
"""

_db.executeAll(sql)
```
