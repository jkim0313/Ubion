```python
# 정규분포

x = np.linspace(-5, 5, 100)
x
```

```python
y = (1 / np.sqrt(2 * np.pi)) * np.exp(-x ** 2 / 2)
y
```

```python
plt.figure(figsize=(10, 6))     
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.title("Normal Distribution without scipy")
plt.legend(["N(0,1)"])
plt.show()
```

```python
# scipy 라이브러리를 이용하여 정규분포 작성

import scipy.stats as stats

x = np.linspace(-5, 5, 100)
y = stats.norm(0, 1).pdf(x)

plt.figure(figsize=(10,6))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend(["N(0,1)"])
plt.show()
```
