### 1번 문제
> For문을 이용하여 구구단을 출력하라
    2단부터 9단까지 구구단을 print를 이용하여 출력하라.

```python
for i in range(2, 10, 1):                   # 2부터 9까지 1씩 증가
   for j in range(1, 10, 1) :               # 1부터 9까지 1씩 증가
        print("%d X %d = %d" %(i, j, i*j))  # 2부터 9까지 구구단 출력
```

### 2번 문제
> 2개의 주사위를 던져서 두 수의 합이 5의 배수인 경우를 출력하라.
1. 2개의 for문을 이용할 것
2. If문을 이용하여 5의 배수 조건을 줄 것

```python
sum = 0
for i in range(1, 7, 1):                        # 주사위 1. 1부터 6까지 1씩 증가
    for j in range(1, 7, 1):                    # 주사위 2.1부터 6까지 1씩 증가
        sum = i + j                             # 주사위 1, 2의 합
        if (sum%5) == 0:                        # 주사위 1, 2의 합을 5로 나눈값이 0일 경우
            print("%d + %d = %d" %(i, j, i+j))  # 주사위 1, 2의 합 출력
            print("%d / 5 = 0" %(i+j))          # 주사위 1, 2의 합을 5로 나눈값 출력
```

### 3번 문제
> Pandas를 이용하여 다음과 같은 데이터프레임을 만들고 데이터프레임을 수정하라.

|   | 0 | 1 |  2 |
|:-:|:-:|:-:|:--:|
| 0 | 1 | 2 |  3 |
| 1 | 7 | 8 |  9 |
| 2 | 4 | 5 |  6 |
| 3 | 3 | 5 | 10 |

1. 컬럼의 이름을 A, B, C로 변경
2. D라는 새로운 컬럼에 shift 함수를 이용하여 C 컬럼의 차를 D 컬럼에 생성(NaN값인 경우에는 0으로 대체)
3. A 컬럼을 기준으로 오름차순 정렬
4. 인덱스의 값을 새로 지정(기존의 인덱스는 삭제)
5. 이 모든 과정은 원본의 데이터를 수정 

```python
import pandas as pd
```

```python
df = pd.DataFrame(                  # 데이터프레임 생성
    {'0' : [1, 7, 4, 3],    
     '1' : [2, 8, 5, 5],
     '2' : [3, 9, 6, 10]
     })

df
```

```python
df.columns = ["A", "B", "C"]         # 컬렴명 A, B, C로 변경

df
```

```python
df["D"] = (df["C"].shift() - df["C"]).fillna(0)     # D 컬럼 생성, C컬럼의 차를 D컬럼에 입력, NaN값인 경우 으로 대체

df                                                  
```

```python
df.sort_values("A", ascending=True, inplace=True)   # A 컬럼을 기준으로 오름차순 정렬

df
```

```python
df.reset_index(drop=True, inplace=True)             # 기존 인덱스는 삭제하고, 인덱스 값 새로 지정

df
```

### 4번 문제
> 2014,2015,2016 csv파일을 로드하여 데이터를 수정하라.

1. 3개의 데이터프레임을 결합하시오 (유니언 결합)
2. 구분 컬럼의 값을 datetime의 형태로 변환하라. 
3. 구분 컬럼에서 년도만 따로 출력하여 파생변수 년도를 생성하라.
4. 년도 별 평균 사망자, 총 사망자를 출력하라. 
5. 년도 별 사고 대비 사망율을 출력하라.

```python
## 2014, 2015, 2016 csv 파일 로드
## 한글 encoding='CP949'로 인코딩

df_2014 = pd.read_csv("C:/ubion/level_test_1_김중헌/level_test_1csv/2014.csv", encoding='CP949')
df_2015 = pd.read_csv("C:/ubion/level_test_1_김중헌/level_test_1csv/2015.csv", encoding='CP949')
df_2016 = pd.read_csv("C:/ubion/level_test_1_김중헌/level_test_1csv/2016.csv", encoding='CP949')
```

```python
## df_2014 출력

df_2014.head()
```

```python
## df_2015 출력

df_2015.head()
```

```python
## df_2016 출력

df_2016.head()
```

```python
dfs_to_union = [df_2014, df_2015, df_2016]              # 유니언 결합할 데이터프레임 리스트

df_union = pd.concat(dfs_to_union, ignore_index=True)   # 데이터 프레임 리스트 유니언 결합, 인덱스 값 새로 지정
df_union
```

```python
df_union["구분"] = df_union["구분"].str.replace("년", "-")  # 년을 datetime 형식을 만들기 위해 '-'로 대체
df_union["구분"] = df_union["구분"].str.replace("월", "")   # 월을 공백으로 대체

df_union["구분"].head()
```

```python
df_union["구분"] = pd.to_datetime(df_union["구분"])     # 구분 컬럼을 datetime 형식으로 타입 변경
df_union["구분"].dt.strftime("%Y%m")                    # 년월 포맷 설정

df_union["구분"].head()
```

```python
df_union.info()         # 구분 컬럼 데이터 타입 변경 확인
```

```python
df_union["년도"] = df_union["구분"].dt.strftime("%Y")   # 년도 컬럼에 구분 컬럼의 년도만 따로 추출해서 입력

df_union.head()
```

```python
df_union.groupby("년도").agg(["mean", "sum"])["사망(명)"]   # 년도별 평균 사망자 수, 총 사망자 수 출력
```

```python
accident = df_union.groupby("년도").sum()["사고(건)"]   # 년도별 총 사고건 수
death = df_union.groupby("년도").sum()["사망(명)"]      # 년도별 총 사망자 수

death_rate = (accident/death)                           # 사고대비 사망율
death_rate
```

### 5번 문제
> 사업용자동차 업종별 위험운전행동(2015).csv 파일을 pandas를 이용하여 read 후 데이터프레임을 수정하라.

1. pandas를 이용하여 csv 파일을 로드한다. (주소는 상대경로로 지정)
2. column의 이름에 공백이 있는 column명이 존재한다. 공백을 제거한다. 
3. pivot_table을 이용하여 월별 업종별 과속 데이터의 평균 값과 합계의 값을 만든다. 
4. matplotlib을 이용하여 pivot_table을 시각화 하여라. (막대그래프를 가로로 표시하라.)

```python
# csv 파일 한글 encoding='CP949'로 인코딩해서 로드
dr_action = pd.read_csv("./level_test_1csv/사업용자동차 업종별 위험운전행동(2015).csv", encoding="CP949")

dr_action.head()
```

```python
dr_action.columns = dr_action.columns.str.replace(" ", "")      # 공백이 있는 컬럼 공백 제거
```

```python
# 월별, 업종별 과속 평균, 합계 피벗테이블 생성
dr_action_pivot = pd.pivot_table(dr_action, index=["월"], columns=["업종"],
                                 values=["과속"], aggfunc=('mean','sum'))

dr_action_pivot
```

```python
## 한글 표시 작업, 시각화

import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # 글씨체 : 맑은 고딕체
matplotlib.rcParams['axes.unicode_minus'] = False

(dr_action_pivot).plot.barh(rot=0, figsize=(15,15))     # 피벗 테이블 바 그래프, 가로로 시각화
```
