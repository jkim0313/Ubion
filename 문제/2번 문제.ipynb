1. transaction 폴더 내에 있는 csv 파일을 모두 데이터프렘으로 불러온다. 
2. transaction_1과 transaction_2를 유니언 결합을 한다. 
3. transaction_detail_1과 transaction_detail_2를 유니언 결합을 한다. 
4. 2,3 번의 데이터프레임의 값을 조인 결합을 한다. 
5. 4번 작업한 데이터프레임을 customer_master, item_master 와 조인 결합한다. 
6. quantity와 item_price의 곱한 값으로 price라는 컬럼을 생성한다. 
7. payment_date컬럼의 데이터형태를 datetime으로 변경한다. 
8. payment_date값의 년 월값만 추출하여 payment_month 컬럼을 생성한다. 
9. groupby를 이용하여 월별, 상품별 price와 quantity의 합계를 출력한다. 


```python
import pandas as pd
```

```python
## 1. transaction 폴더 내에 있는 csv 파일을 모두 데이터프렘으로 불러온다. 

trans_1 = pd.read_csv("C:/ubion/문제/1회차/num_2/transaction_1.csv")
trans_2 = pd.read_csv("C:/ubion/문제/1회차/num_2/transaction_2.csv")
detail_1 = pd.read_csv("C:/ubion/문제/1회차/num_2/transaction_detail_1.csv")
detail_2 = pd.read_csv("C:/ubion/문제/1회차/num_2/transaction_detail_2.csv")
i_mast = pd.read_csv("C:/ubion/문제/1회차/num_2/item_master.csv")
c_mast = pd.read_csv("C:/ubion/문제/1회차/num_2/customer_master.csv")
```

### concat()

1. concat 함수는 데이터프레임을 결합하는 함수
2. axis 0, 1 설정시 행, 열 추가(기본값은 행을 추가)
3. 단순하게 행이나 열을 합치는기능.

```python
## 2. transaction_1과 transaction_2를 유니언 결합을 한다. 

trans_dfs = [trans_1, trans_2]

trans_union = pd.concat(trans_dfs, ignore_index=True)
trans_union
```

```python
## 3. transaction_detail_1과 transaction_detail_2를 유니언 결합을 한다. 

detail_dfs = [detail_1, detail_2]

detail_union = pd.concat(detail_dfs, ignore_index=True)
detail_union
```

```python
## 4. 2,3 번의 데이터프레임의 값을 조인 결합을 한다. 

td_join = pd.merge(trans_union, detail_union, how = "inner", on = "transaction_id")

td_join
```

```python
## 5. 4번 작업한 데이터프레임을 customer_master, item_master 와 조인 결합한다.

tdc_join = pd.merge(td_join, c_mast, how="inner", on="customer_id")

tdc_join
```

```python
tdic_join = pd.merge(tdc_join, i_mast, how="inner", on="item_id")

tdic_join
```

```python
## 6. quantity와 item_price의 곱한 값으로 price라는 컬럼을 생성한다. 

tdic_join['price'] = tdic_join['quantity'] * tdic_join['item_price']

tdic_join
```

```python
## 7. payment_date컬럼의 데이터형태를 datetime으로 변경한다. 
## 8. payment_date값의 년 월값만 추출하여 payment_month 컬럼을 생성한다. 

tdic_join.info()
```

```python
tdic_join['payment_date'] = pd.to_datetime(tdic_join['payment_date'])

tdic_join['payment_date'].dt.strftime("%Y%m")
tdic_join.info()
```

```python
tdic_join.head()
```

```python
## 9. groupby를 이용하여 월별, 상품별 price와 quantity의 합계를 출력한다. 

tdic_join.groupby(['payment_month', 'item_name']).sum()[['price','quantity']]
```
