6. 데이터의 결측치가 존재하는지 확인한다. 
7. item_price 컬럼의 결측치의 값을 해당하는 item_name의 맥스값으로 치환하여 준다. 

```python
import pandas as pd
import numpy as np
```

```python
## 1. num_1 폴더의 uriage 파일을 pandas를 이용하여 데이터프레임으로 불러온다. 

uri_df = pd.read_csv("C:/ubion/문제/1회차/num_1/uriage.csv")

uri_df
```

```python
uri_df.info()
```

```python
## 2. purchase_date 컬럼의 데이터 타입을 datetime을 변경한다. 

uri_df['purchase_date'] = pd.to_datetime(uri_df['purchase_date'])
```

```python
## 3. item_name 컬럼의 값을 대문자로 변경한다.
uri_df['item_name'] = uri_df['item_name'].str.upper()

uri_df
```

```python
## 4. item_name 컬럼의 공백을 지워준다.
uri_df['item_name'] = uri_df['item_name'].str.replace(" ", "")
```

```python
uri_df
```

```python
## 5. item_name 컬럼을 기준으로 오름차순 정렬한다.
uri_df = uri_df.sort_values(by='item_name', ascending=True)
```

```python
# 7. item_price 컬럼의 결측치의 값을 해당하는 item_name의 맥스값으로 치환하여 준다. 

# uri_df["item_price"].apply(lambda x : item_max if x == np.NaN else x)

flg_is_null = uri_df["item_price"].isna()

# item_price의 값이 NaN인 item_name들의 중복값을 삭제한 arrary형태
uri_df.loc[flg_is_null, "item_name"].unique()
```

```python
# ~연산자는 bool형태의 데이터를 반전 시켜준다.

uri_df.loc[(~flg_is_null) & (uri_df["item_name"] == "상품 A"), "item_price"].max()
```

```python
for trg in list(uri_df.loc[flg_is_null, "item_name"].unique()):
    price = uri_df.loc[(~flg_is_null) & (uri_df["item_name"] == trg), "item_price"].max()
    # print(trg, price)
    uri_df["item_price"].loc[(flg_is_null) & (uri_df["item_name"] == trg)] = price
```
