```python
df_3 = sales_df_2.copy()
```

```python
## drop(특정값, axis=n)

df_3 = df_3.drop(['발송일자'], axis=1) 
df_3
```

```python
df_3 = df_3.drop([1], axis=0)

df_3
```

```python
## inplace 속성은 기준이 되는 데이터프레임을 교체를 할지 지정하는 속성
## 기본값은 False. True로 변경을 하게되면 기준이 되는 데이터프레임을 수정

df_3.sort_values('주문일자', inplace=True)

df_3
```

```python
df_3.drop('주문ID', axis = 1, inplace = True)

df_3
```

```python
## reset_index(drop=True, inplace=True)
## 기존의 인덱스 값 유지, 인덱스 재정렬

df_3.reset_index(drop=True, inplace=True)

df_3
```

```python
df_3 = df_3.sort_values("권역")

df_3.sort_values("권역", inplace=True)
```

```python
df_3['총이윤'] = df_3['총수익'] - df_3['총비용']

df_3.head()
```

```python
df_3.drop(['총수익', '총비용', '총이윤'], axis=1, inplace=True)

df_3
```

df_3 데이터프레임에서 총수익, 총비용, 총이윤이라는 파생변수를 생성.

```python
df_3['총수익'] = df_3['판매단위'] * df_3['단가']
df_3['총비용'] = df_3['판매단위'] * df_3['원가']
df_3['총이윤'] = df_3['총수익'] - df_3['총비용']
```
